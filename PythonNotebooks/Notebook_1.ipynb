```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
full_df = pd.read_csv("train.csv")
full_df_group = full_df.groupby(by=["customer_ID"], sort=False) # .agg(lambda x: )
df_list = []
pol_change = False
chan_shpt = 0
customer_id = []
change_lst = []
change_shpt_lst = []
policy_lst = []
cost_lst = []
time_diff = []
for grp, data in full_df_group :
    temp_data = data
    #print(grp)
    x = list(data["C_previous"])
    p = list(data["cost"])
    q = list(data["policy"])
    timeObj = []
    date_format = "%H:%M"
    t = list(data["time"])
    frst_time = dt.strptime(t[0], date_format)
    lst_time = dt.strptime(t[len(t) - 1], date_format)
    diff = lst_time - frst_time
    time_diff.append(int(diff.seconds / (60*60)))
    cost_lst.append(p[len(x) - 1])
    policy_lst.append(q[len(q) - 1])
    z = x[len(x) - 1]
    pol_previous = x[len(x) - 1] # final value of C_previous
    pol_change = 0 # didn't change
    change_shpt = 0
    for index,row in data.iterrows():
        if(row["C_previous"] == z):
            break
        else:
            pol_change = 1 # changed
            break
    if(pol_change == 1):
        pol = x[0]
        for index,row in temp_data.iterrows():
            c = row["C_previous"]
            change_shpt = change_shpt + 1
            if(str(pol) != str(c)):
                break
       
    customer_id.append(grp)
    change_lst.append(pol_change)
    change_shpt_lst.append(change_shpt)

result = pd.DataFrame({'customer_ID': customer_id, 'policy_change':change_lst, 'change_shpt':change_shpt_lst, 'policy': policy_lst, 'cost': cost_lst, 'time_diff': time_diff})

result.to_csv("session.csv")
```
